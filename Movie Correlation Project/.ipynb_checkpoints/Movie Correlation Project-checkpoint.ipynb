# Data Cleaning

### In order to begin dissecting and diving into this dataset, we should first clean it!

Let's see if there is any missing data.

```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())*100
    print('{} - {}% missing'.format(col, pct_missing))
```

It looks like there are quite a few rows with at least 1 null value. Let's take a closer look.

```python
withnulls = df[df.isnull().any(axis=1)]
withnulls.shape
```

```python
df.isnull().sum()
```

It appears that a majority of the null values in this dataset are in the 'budget' column. Since this is such an important column for the purpose of this project, the safest bet is to remove rows where there are null values, as attempting to impute values based on trends or averages can yield inaccurate results.

```python
original_size = df.shape
df = df.dropna()
df.shape
```

```python
original_size[0] - withnulls.shape[0]
```

Next, let's take a look at our data types to see if we need to change anything to get our data ready for use.

```python
df.dtypes
```

Seems like things are good. But, they can be better. Some of these types, such as budget and votes can be turned into ints as it is unnecessary for them to be in float format.

```python
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df['votes'] = df['votes'].astype('int64')
df['runtime'] = df['runtime'].astype('int64')
```

```python
df.head()
```

A deeper dive into the data reveals that the 'year' column contains some inaccuracies, stating a year the is not the release year. An example of this is the movie 'Superman II' (index 8), where that row states that the release date is June 19, 1981 whereas the year column lists 1980.

```python
df.head(10)
```

To correct this, we extract the year straight from the strings in the 'released' column using a regex pattern. We use the pattern 
'([0-9]{4})' to indicate that we're looking for a length 4 string of only numbers 0-9.

```python
df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
df['yearcorrect'] = df['yearcorrect'].astype('int64')
df.head()
```

```python
#As seen here, there were several hundred entries where this error was made.
mismatched = df.loc[df['year'] != df['yearcorrect']]
mismatched.shape
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False).head(10)
```

```python
pd.set_option('display.max_rows', None)
```

Next, let's quickly drop any duplicates.

```python
df.drop('year', axis=1, inplace=True)
df_shape = df.shape
df = df.drop_duplicates()
print(df_shape == df.shape)
```

# Data Analysis

Now that we cleaned up our data, we can start the analysis.
### 1. Budget and Gross Correlation: 
I predict that there is a high correlation between budget and gross values, where the higher the budget for the movie, the higher the movie will earn (gross).

```python
plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget for Film')
plt.ylabel('Gross Earnings')
plt.show()
```

```python
sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

As seen in the plots above, there seems to be SOME positive correlation between the budget and gross of the films. From the seaborn plot, there is a line trending upwards, indicating that generally speaking, the higher the budget for a movie, the more likely it is that it will earn more.

```python
df.corr()
```

Looking at the stats above, if we look only at the numerical values in the data, we can see that the highest correlation lies between budget and gross, having a score of +0.740247. The next highest would be gross and votes at +0.614751, then votes and score at +0.474256 (although the votes and the score correlation may be a bit obvious).

Let's take a look at some other correlation methods and see if they agree.

```python
df.corr(method='kendall')
```

This method says that the strong correlation lies between the gross and the votes!

```python
df.corr(method='spearman')
```

Same with this one! What's going on here?

According to these methods, it seems that the higher the number of votes for a movie, the higher the gross. If you think about it, it makes some sense - People who vote for a movie are likely to have already seen it, which implies that they bought a ticket. If more people are voting, that means more people are buying tickets for the movie, hence the higher gross.

```python
#Another view
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

### 2. Company Correlation
Now, let's see if company has a significant impact on the gross of the movie! There are a few famous companies such as Disney or Lucasfilms that I think may have an impact on the sales of the movie.

First, we numerize every column.

```python
df_numerized = df.copy()

for col in df_numerized.columns:
    if df_numerized[col].dtype == 'object':
        df_numerized[col] = df_numerized[col].astype('category')
        df_numerized[col] = df_numerized[col].cat.codes
        
df_numerized.head()
```

Now we can take a look at its correlation matrix.

```python
correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
df_numerized.corr()
```

```python
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
sorted_pairs[(abs(sorted_pairs) > 0.5)]
```

According to the above data, it seems like not much besides budget and votes have a significant impact on gross, not even company. However, there is some other data worth noting above, such as there seems to be a negative correlation between budget/gross and genre.
